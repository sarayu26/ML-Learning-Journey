# Breast Cancer Binary Classification Using Logistic Regression

This notebook performs binary classification on the Breast Cancer dataset using logistic regression. The task is to classify tumors as malignant or benign based on various features extracted from cell nuclei images. The notebook covers data loading, exploration, preprocessing, model training, evaluation, and visualization.


```python
# Import necessary libraries for data manipulation, visualization, and machine learning
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import machine learning tools
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report, roc_auc_score, roc_curve,auc

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load Breast Cancer dataset from sklearn and convert to pandas DataFrame
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

df = pd.DataFrame(data = np.c_[cancer.data, cancer.target],columns = list(cancer.feature_names)+['target'])
```

## Dataset Overview

The dataset contains 569 samples with 30 numeric features each describing characteristics of cell nuclei, along with a binary target variable (`0` for malignant, `1` for benign tumors).


```python
# Show first few rows of dataset for initial inspection
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Prepare Data for Modeling

Split the data into features (`X`) and target labels (`y`). Then split into training and test sets (80% train, 20% test).

Because logistic regression benefits from standardized features, apply StandardScaler to scale input features to mean 0 and variance 1.


```python
x = df.drop('target',axis = 1)
y= df['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

## Train Logistic Regression Model

We use sklearn's LogisticRegression to train a binary classifier on the training data.


```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
cr = classification_report(y_test,y_pred)
print(cr)
```

```python
y_pred_prob = model.predict_proba(x_test)[:,1]
roc_auc =roc_auc_score(y_test,y_pred_prob)
fpr, tpr, _ = roc_curve(y_test,y_pred_prob)
```

```python
plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,color = 'red',lw=2,label='ROC curve(area=%0.2f)' % roc_auc)
plt.plot([0,1],[0,1],color='navy',lw=2,linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Chareteristics')
plt.legend(loc='lower right')
```

## Conclusion

The logistic regression classifier achieves high accuracy and performs well in distinguishing malignant from benign tumors. Visualization of the confusion matrix and ROC curve supports these findings.
