# California Housing Price Prediction Using Linear Regression

This notebook explores the California housing dataset, applies linear regression to predict housing prices, and evaluates the model's performance using various error metrics.


```python
# Import necessary libraries for numerical computations, data manipulation, and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import dataset and machine learning tools from scikit-learn
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error , r2_score
```

```python
# Load the California housing dataset
cal_housing=fetch_california_housing()
```

```python
# Create a DataFrame from the dataset with feature names and target price
data = pd.DataFrame(data=np.c_[cal_housing['data'],cal_housing['target']],columns =cal_housing['feature_names']+['target'])
# Create a DataFrame from the dataset with feature names and target price
data.head()
```

```python
# Separate independent variables (features) and dependent variable (target)
x=data.drop('target',axis=1)
y=data['target']
```

```python
# Split data into training and testing sets (70% train, 30% test)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# Initialize and train a Linear Regression model using training data
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
# Predict housing prices on the test set
y_pred = model.predict(x_test)
```

```python
# Calculate evaluation metrics to assess model performance
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)

print(f'MAE: {mae:2f}')
print(f'MSE: {mse:2f}')
print(f'R2: {r2:2f}')
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test,y_pred,color='red',alpha=0.3)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],'k--',lw=2)
plt.xlabel('actual values')
plt.ylabel('predicted values')
plt.title('actual vs predicted values')
plt.show()
```

## Conclusion

The linear regression model shows reasonable performance on the California housing dataset with an R² score indicating the proportion of variance explained by the model. However, there is room for improvement by exploring more complex models or feature engineering.
